```python
# Usefull functions




def syllable_count(word):
    word = word.lower()
    count = 0
    vowels = "aeiou"
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
    if word.endswith("e"):
        count -= 1
    if count == 0:
        count += 1
    return count


def remove_stopwords(words):
    new_words = []
    for word in words:
        if word not in my_stop_words:
            new_words.append(word)
    return new_words



def removePunctuation(content):
    content.translate(str.maketrans('', '', string.punctuation))
    return content


def syllable_per_word(content):
    word=content.replace(' ','')
    syllable_count=0
    for w in word:
          if(w=='a' or w=='e' or w=='i' or w=='o' or w=='y' or w=='u' or w=='A' or w=='E' or w=='I' or w=='O' or w=='U' or w=='Y'):
                syllable_count=syllable_count+1
    return syllable_count/len(content.split())
```

## Extracting text from all the url


```python
 ## header is used to give user-access, if not using we will get error 406
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
li = []
for i in range(0,len(df)):
    try:
        
        j = df.iloc[i].values
        k=j[1]
        print('\nj[0]',j[0])
        print(k)


        ## getting the content from url and storing it in page
        page = requests.get(k,headers=headers)


        ## since page.content is not so readable there we will make it readable  
        soup=BeautifulSoup(page.content,'html.parser') #parsing url text
        print(page)


        ## We will extract text part
        #content=soup.findAll(attrs={'class':'td-post-content'})
        content = soup.find('div',{'class':'td-post-content'})        
        content=content.text
        content=content.replace('\n',"  ")


        ## similarly with title
        #title = soup.findAll(attrs={'class':'entry-title'})
        title = soup.find('h1',{'class':'entry-title'})
        title = title.text.replace('\n',' ').replace('/',' ')
        print(title)




        ## saving the content in text file
        URL_ID=str(j[0])+'.txt'
        #print(URL_ID)
        f=io.open(URL_ID,'w', encoding="utf-8")
        f.writelines(content)
        f.close()




        content = removePunctuation(content)
        #print(content)


        ## total words in content
        text_tokens = word_tokenize(content)
        print("Total words in content: ",len(text_tokens))

        ## removing stopwords
        newContent = remove_stopwords(text_tokens)
        #print(newContent)

        print('After removing stop words: ',len(newContent))
        #word_cloud(newContent)

        
        ## OBJECTIVES
        
        ## getting positive,negative words
        positive_score=0
        for i in newContent:
              if(i.lower() in pos):
                    positive_score+=1
        print('postive score=', positive_score)
        negative_score=0
        for i in newContent:
              if(i.lower() in neg):
                    negative_score+=1
        print('negative score=', negative_score)
        
        
        
        ## porality score
        polarityScore = (positive_score-negative_score)/((positive_score+negative_score)+0.000001)
        print('polarityScore: ',polarityScore)
        
        
        ## subjective score
        subjectiveScore = (positive_score+negative_score)/(len(newContent)+0.000001)
        print('subjectiveScore: ',subjectiveScore)
        
        
        #AVG SENTENCE LENGTH
        AVG_SENTENCE_LENGTH = len(content.replace(' ',''))/len(re.split(r'[?!.]', content))
        print('AVG_SENTENCE_LENGTH: ',AVG_SENTENCE_LENGTH)
        
        
        # complex word
        count=0
        for word in text_tokens:
            c=syllable_count(word)
            if c>2:
                count+=c
                
        print('Complex word count: ',count)

        complexPercentage=count/len(text_tokens)
        print('complexPercentage: ',complexPercentage)
        
        

        # fog index
        fog=0.4*(AVG_SENTENCE_LENGTH+complexPercentage)
        print('fog: ',fog)

        

        # Avg number of word per sentence
        AVG_NUMBER_OF_WORDS_PER_SENTENCE = [len(l.split()) for l in re.split(r'[?!.]', content) if l.strip()]
        AVG_NUMBER_OF_WORDS_PER_SENTENCE=sum(AVG_NUMBER_OF_WORDS_PER_SENTENCE)/len(AVG_NUMBER_OF_WORDS_PER_SENTENCE)
        print('Avg number of words per sentence',AVG_NUMBER_OF_WORDS_PER_SENTENCE)

        
        
        # Word count
        wordCount=len(newContent)
        print('Word count: ',wordCount)
        
        

        #syllable per word
        syllablePerWord = syllable_per_word(content)
        print('Syllable per word: ',syllablePerWord)



        # personal pronoun
        pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
        pronouns = pronounRegex.findall(content)
        personal_pronoun = len(pronouns)
        print('pronouns: ',personal_pronoun)
        
        
        
        # avg word length
        Average_Word_Length=len(content.replace(' ',''))/len(content.split())
        print('Average_Word_Length: ',Average_Word_Length)
        


        
    except Exception as e:
        print(f"Error occured in {k}")
        print(str(e))

    
    
    data={'POSITIVE SCORE':positive_score,'NEGATIVE SCORE':negative_score,'POLARITY SCORE':polarityScore,'SUBJECTIVITY SCORE':subjectiveScore,'AVG SENTENCE LENGTH':AVG_SENTENCE_LENGTH,'PERCENTAGE OF COMPLEX WORDS':complexPercentage,
          'FOG INDEX':fog,'AVG NUMBER OF WORDS PER SENTENCE':AVG_NUMBER_OF_WORDS_PER_SENTENCE,'COMPLEX WORD COUNT':count,
          'WORD COUNT':wordCount,'SYLLABLE PER WORD':syllablePerWord,'PERSONAL PRONOUNS':personal_pronoun,'AVG WORD LENGTH':Average_Word_Length}

    li.append(data)
    

    
        
```
